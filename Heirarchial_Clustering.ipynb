```python
dataset = pd.read_csv('Wholesale customers data.csv')
dataset
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(dataset)
data_scaled = pd.DataFrame(data_scaled,columns=dataset.columns)
data_scaled.head()
x=data_scaled
x.head()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method='complete'))
plt.title("Dendrogram")
plt.xlabel('customers')
plt.ylabel('Euclidian distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='complete')
y_hc = hc.fit_predict(x)
```

```python
plt.scatter(x['Milk'],x['Detergents_Paper'],c=hc.labels_)
plt.title('Clusters')
plt.xlabel('unit purchased')
plt.ylabel('clients')
plt.show()
```

```python
plt.scatter(x['Milk'],x['Grocery'],c=hc.labels_)
plt.title('Clusters')
plt.xlabel('unit purchased')
plt.ylabel('clients')
plt.show()
```
